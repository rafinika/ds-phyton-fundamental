## *What are the odds that you'll reach 60 steps high on the Empire State Building?*


Rules:
1.   Roll a dice, when you land on:


    *   1 or 2 -> backward 1 step
    *   3, 4 or 5 -> forward 1 step
    *   6 -> roll one more time, any number you land on, go forward those amount

2.   Can't go below step 0
3.   0.1% chance of falling down the stairs


 Can you reach step 60?






```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7064)
```

```python
# Simulate random walk 10 times
all_walks = []
for i in range(10) :
  random_walk =[0]
  for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)
    if dice <= 2: # Roll on 1 or 2
      step = max(0, step - 1) # Can't go bellow 0
    elif dice <= 5:
      step = step +1
    else:
      step = step + np.random.randint(1,7)

    # Implement clumsiness
    if np.random.rand() <= 0.001:
      step = 0

    random_walk.append(step)
  all_walks.append(random_walk)

# Visualize with plot
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.xlabel(" dice roled")
plt.ylabel(" step reached")
plt.show()
```

```python
# Simulate random walk 1000 times
all_walks = []
for i in range(1000) :
  random_walk =[0]
  for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)
    if dice <= 2:
      step = max(0, step - 1)
    elif dice <= 5:
      step = step +1
    else:
      step = step + np.random.randint(1,7)
    if np.random.rand() <= 0.001:
      step = 0
    random_walk.append(step)
  all_walks.append(random_walk)

# Get the last step
np_aw_t = np.transpose(np.array(all_walks))
ends = np_aw_t[-1]

# Plot histogram of ends
plt.hist(ends)
plt.xlabel("step reached at the end")
plt.ylabel("number of walk")
plt.show()
```

```python
result = np.sum(ends >= 60) / 1000
print(f"Probability you ended above 60 steps: {result}")
```
